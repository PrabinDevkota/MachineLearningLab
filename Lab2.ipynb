```python
#Defining functions for modularization
def mean(data):
  sum = 0;
  i = 0;
  for num in data:
    sum = sum + num
    i = i+1
  return sum/i


def variance(data):
  n = len(data)
  if n == 0:
    return 0

  # Calculate the mean
  mean_val = mean(data) # Using the previously defined mean function

  # Calculate the sum of squared differences from the mean
  sum_sq_diff = 0
  for num in data:
    sum_sq_diff += (num - mean_val)**2

  return sum_sq_diff / n
```

```python
#Cost per unit of the product
# Extract quantities of products (Candies, Mangoes, Milk Packets) as matrix A
A = Y[:, :3]
# Extract total payment as vector b
b = Y[:, 3]

# Calculate the pseudo-inverse of A
A_inv = lin.pinv(A)

# Calculate the cost per unit of each product
cost_per_unit = A_inv @ b

cost_per_unit
```

```python
# Remove N/A data present in price column
price_data = stock['Price'].str.replace(',', '').to_numpy(dtype=float)
price_data = price_data[~np.isnan(price_data)]
price_data
```

```python
int(np.mean(price_data))
```

```python
# My own mean calculation
print(mean(price_data))
```

```python
# My own variance calculation
print(variance(price_data))
```

```python
# Filter for 'Wed' (Wednesday) data
wednesday_stock = stock[stock['Day'] == 'Wed']

# Extract and clean 'Price' data for Wednesdays
wednesday_price_data = wednesday_stock['Price'].str.replace(',', '').to_numpy(dtype=float)
wednesday_price_data = wednesday_price_data[~np.isnan(wednesday_price_data)]

# Calculate the mean of Wednesday prices using the custom mean function
mean_wednesday_price = mean(wednesday_price_data)

print(f"Mean price on Wednesdays: {mean_wednesday_price:.2f}")
```

```python
# Filter for April  data
april_stock = stock[stock['Month'] == 'Apr']


april_price_data = april_stock['Price'].str.replace(',', '').to_numpy(dtype=float)
april_price_data = april_price_data[~np.isnan(april_price_data)]

april_mean_price = mean(april_price_data)

print(f"Mean price in April: {april_mean_price:.2f}")
```

```python
# Extract 'Chg%' column
chg_percent_data = stock['Chg%']

# Clean the data: remove '%' sign, convert to float, and drop NaN values
# Apply a lambda function to remove '%' and convert to float, coercing errors to NaN
chg_percent_numeric = chg_percent_data.apply(lambda x: float(x.strip('%')) if isinstance(x, str) else np.nan)
chg_percent_numeric = chg_percent_numeric.dropna()

# Count the number of negative values (losses) using a lambda function
negative_changes = chg_percent_numeric[chg_percent_numeric.apply(lambda x: x < 0)]

# Calculate the probability of a loss
probability_of_loss = len(negative_changes) / len(chg_percent_numeric)

print(f"Total valid 'Chg%' entries: {len(chg_percent_numeric)}")
print(f"Number of negative 'Chg%' entries (losses): {len(negative_changes)}")
print(f"Probability of making a loss: {probability_of_loss:.4f}")
```

```python
# Probability of making profit on wednesday
wed_chg = stock[stock['Day'] == 'Wed']['Chg%']
wed_chg = wed_chg.apply(lambda x: float(x.strip('%')) if isinstance(x, str) else np.nan)
wed_chg.to_numpy()
change_count = wed_chg.size
pos_change = wed_chg[wed_chg > 0]
pos_change_count = pos_change.size

neg_change = wed_chg[wed_chg < 0]
neg_change_count = neg_change.size
neg_change_count

probability_of_loss = neg_change_count / change_count
probability_of_loss

probability_of_profit = 1 - probability_of_loss
probability_of_profit
```
