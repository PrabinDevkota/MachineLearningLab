```python
# Function to count vowels and consonants
def count_vowels_consonants(input_string):
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0

    for character in input_string:
        if character.isalpha():
            if character in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    return vowel_count, consonant_count


def main():
    input_string = input("Enter a string: ")

    vowels, consonants = count_vowels_consonants(input_string)

    print("Number of vowels:", vowels)
    print("Number of consonants:", consonants)


if __name__ == "__main__":
    main()
```

```python
# Function to multiply two matrices
def multiply_matrices(matrix_a, matrix_b):
    rows_a = len(matrix_a)
    columns_a = len(matrix_a[0])
    columns_b = len(matrix_b[0])

    result = []

    for i in range(rows_a):
        row = []
        for j in range(columns_b):
            value = 0
            for k in range(columns_a):
                value += matrix_a[i][k] * matrix_b[k][j]
            row.append(value)
        result.append(row)

    return result


def main():
    # Hardcoded Matrix A
    matrix_a = [
        [1, 2, 3],
        [4, 5, 6]
    ]

    # Hardcoded Matrix B
    matrix_b = [
        [7, 8],
        [9, 10],
        [11, 12]
    ]

    # Check if multiplication is possible
    if len(matrix_a[0]) != len(matrix_b):
        print("Error: Matrix multiplication is not possible.")
    else:
        result = multiply_matrices(matrix_a, matrix_b)

        print("Matrix A:")
        for row in matrix_a:
            print(row)

        print("\nMatrix B:")
        for row in matrix_b:
            print(row)

        print("\nProduct Matrix (A × B):")
        for row in result:
            print(row)


if __name__ == "__main__":
    main()
```

```python
# Function to find common elements
def find_common_elements(list_one, list_two):
    common_elements = list(set(list_one) & set(list_two))
    return common_elements


def main():
    list_one = list(map(int, input("Enter first list elements: ").split()))
    list_two = list(map(int, input("Enter second list elements: ").split()))

    common = find_common_elements(list_one, list_two)

    print("Common elements:", common)
    print("Number of common elements:", len(common))


if __name__ == "__main__":
    main()
```

```python
# Function to read a matrix
def input_matrix(rows, columns):
    matrix = []

    for i in range(rows):
        row = list(map(int, input().split()))
        matrix.append(row)

    return matrix


# Function to find transpose
def transpose_matrix(matrix):
    rows = len(matrix)
    columns = len(matrix[0])

    transpose = []

    for j in range(columns):
        row = []
        for i in range(rows):
            row.append(matrix[i][j])
        transpose.append(row)

    return transpose


def main():
    rows = int(input("Enter number of rows: "))
    columns = int(input("Enter number of columns: "))

    print("Enter matrix:")
    matrix = input_matrix(rows, columns)

    transpose = transpose_matrix(matrix)

    print("Transpose Matrix:")
    for row in transpose:
        print(row)


if __name__ == "__main__":
    main()
```

```python
import random
import statistics


# Function to generate random numbers
def generate_random_numbers(count, lower_limit, upper_limit):
    numbers = []

    for i in range(count):
        numbers.append(random.randint(lower_limit, upper_limit))

    return numbers


# Function to calculate statistics
def calculate_statistics(numbers):
    mean_value = statistics.mean(numbers)
    median_value = statistics.median(numbers)
    mode_value = statistics.multimode(numbers)

    return mean_value, median_value, mode_value


def main():
    numbers = generate_random_numbers(100, 100, 150)

    mean_value, median_value, mode_value = calculate_statistics(numbers)

    print("Generated Numbers:")
    print(numbers)

    print("\nMean:", mean_value)
    print("Median:", median_value)
    print("Mode:", mode_value)


if __name__ == "__main__":
    main()
```
